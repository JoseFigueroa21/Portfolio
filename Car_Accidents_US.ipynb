```python
def q1(data):
    most_accidents_month_name = pd.to_datetime(data['Start_Time']).dt.month_name().mode()[0]
    print("(1) In what month were there more accidents reported?", most_accidents_month_name)
```

```python
def q2(data):
    state_with_most_accidents_2020 = data.loc[data['Start_Time'].str.startswith('2020'), 'State'].value_counts().idxmax()
    print("(2) What is the state that had the most accidents in 2020?", state_with_most_accidents_2020)
```

```python
def q3(data):
    top_5_cities_2019_ca = data.loc[(data['Start_Time'].str.startswith('2019')) & (data['State'] == 'CA'), 'City'].value_counts().head(5)
    print("(3) What are the 5 cities that had the most accidents in 2019 in California?")
    print(top_5_cities_2019_ca.to_string(header=False))
```

```python
def q4(data):
    average_humidity = data.loc[(data['Severity'] == 4) & (data['Start_Time'].str.startswith('2021')), 'Humidity(%)'].mean()
    average_temperature = data.loc[(data['Severity'] == 4) & (data['Start_Time'].str.startswith('2021')), 'Temperature(F)'].mean()
    print("(4) What was the average humidity and average temperature of all accidents of severity 4 that occurred in 2021?")
    print("Average Humidity: {:.2f}".format(average_humidity))
    print("Average Temperature: {:.2f}".format(average_temperature))
```

```python
def q5(data):
    max_visibility = data.loc[(data['Severity'] == 2) & (data['State'] == 'NH'), 'Visibility(mi)'].max()
    print("(5) What was the maximum visibility of all accidents of severity 2 that occurred in the state of New Hampshire? {:.2f}".format(max_visibility))
```

```python
def q6(data):
    severity_counts = data.loc[data['City'] == 'Bakersfield', 'Severity'].value_counts()
    print("(6) How many accidents of each severity were recorded in Bakersfield?")
    print(severity_counts.to_string(header=False))
```
